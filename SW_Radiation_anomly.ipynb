```python
from mpl_toolkits.basemap import Basemap
import xarray as xr
import matplotlib.pyplot as plt
import PIL
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
data_2000_2020 = xr.open_dataset('/content/CERES_SYN1deg-Month_Terra-Aqua-MODIS_Ed4.1_Subset_200003-202003 (1).nc')
data_2019 = xr.open_dataset('/content/CERES_SYN1deg-Month_Terra-Aqua-MODIS_Ed4.1_Subset_201901-201912.nc')
data_2000_2020
```

```python
data_2000_2020.variables.keys()
```

```python
lats = data_2000_2020['lat'][:]
```

```python
lons = data_2000_2020['lon'][:]
```

```python
time1 = data_2000_2020['time'][:]
time2 = data_2019['time'][:]
time1[10]
```

```python
sw1 = data_2000_2020['toa_sw_all_mon'][:]
sw2 = data_2019['toa_sw_all_mon'][:]
```

```python
idx = np.arange(10,242,12)
idx
```

```python
time1[10]
#time1[226]
```

```python
sw_2019 = np.array(sw1[226])
sw_2019.shape
```

```python
type(sw1)
```

```python
sw11 = np.array(sw1)
sw11_jan = np.array([sw11[i] for i in idx])
sw11_jan.shape
```

```python
sw11_mean_jan = np.mean(sw11_jan, axis=0)
sw11_mean_jan.shape
```

```python
anomaly_jan = sw_2019 - sw11_mean_jan
anomaly_jan
```

```python
mp =Basemap()
```

```python
lon,lat = np.meshgrid(lons,lats)
x,y = mp(lon,lat)
```

```python
plt.subplots(figsize=(15,6))
c_scheme = mp.pcolor(x,y, anomaly_jan, latlon=True, cmap = 'jet', vmin=-30, vmax=30)
mp.drawcoastlines()
mp.drawstates()
mp.drawcountries()
mp.shadedrelief()
cbar = mp.colorbar(c_scheme, location= 'right', pad= '10%')
mp.drawparallels([-90,-60,-30,0,30,60,90], labels=[1,0,0,0])
mp.drawmeridians([0,60,120,180,240,300,360], labels=[0,0,0,1])
plt.ylabel('Latitude', fontweight= 'bold')
plt.xlabel('Longitude', fontweight= 'bold')
plt.title('Jan 2019 Toa SW flux anamoly', fontweight='bold')
plt.savefig('jan_anomaly_sw.jpg', dpi=400)
```
